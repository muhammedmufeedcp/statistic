#Qualitiative data

```python
#qualitatitive data example (text and catagories)
#creating lists with qualitative data

products=["apple","banana","cherry","apple","banana"]
reviews=["good","avarage","excellent","good","poor"]

#displaying the qualitative data list

print("qualitative data:")
print("products:",products)
print("reviews:",reviews)
```

```python
#showing data types

print("data types of qualitative data:")
print("products:",type(products))
print("reviews:",type(reviews))
```

```python
#showing data types

print("data types of qualitative data:")
print("products:",type(products[0]))
print("reviews:",type(reviews[0]))
```

# quatitative data

```python
#quatitative data examle (integers and floats)
#creating list with quatitative data

product_index=[101,102,103,104,105,] #index number
prices=[0.99,0.5,1.99,0.99,0.5]
quantities=[10,20,15,10,5]

#displaying the quatitative data list

print("quatitative data")
print("products index:",product_index)
print("prices:",prices)
print("quantitative:",quantities)
```

```python
#showing data types

print("data types of quantitative data:")
print("products_index:",type(product_index))
print("prices:",type(prices))
print("quantitative:",type(quantities))
```
