# mean

mean(avarage):the sum of all the value divided by the number of value

```python
(10+15+20+25+30+35+40+45+50+55+60)/11
```

```python
# mean example

import numpy as np
data=[1,2,3,4,5]
mean_value=np.mean(data)
print("mean (avarage):")
print("data:",data)
print("mean is :",mean_value)
```

```python
#mean identify another way using statistics method

import statistics as stati
print(stati.mean([50,60,40,20]))
```

# median

```python
# median example

import numpy as np
data=[1,2,3,4,5]
median_value=np.median(data)
print("median (avarage):")
print("data:",data)
print("median is :",median_value)
```

```python
# median identify another way using statistics method

import statistics as stati
print(stati.median([50,60,40,20]))
```

if the number of data values is odd,it returns the exact value. if the number of data values is even, it returns the avarage of the two middle values.

```python
print(stati.median([1,3,5,7,9,11,13]))
# the number of data values is odd
```

```python
print(stati.median([1,3,5,7,9,11,]))
# the number of the data value is even
# 5+7/2=6
```

```python
print(stati.median([9,6,8,1,4,2]))
#1,2,4,6,8,9 after sorting
```

# mode

the most frequently occuring value in dataset


mudule "numpy" has no atribute "mode"



```python
# mode using statistic module
mode_value = stati.mode(data)

print("mode is :",mode_value)

# this data has not repeating value
```

if all values occur with the same frequency,statistic.mode()returns the first value in the order of the dataset,which is 1 in this case

```python
data_with_mode=[1,2,3,3,3,3,4,5,6]
mode_value=stati.mode(data_with_mode)

print("mode is:",mode_value)
```

## mode for qualitative data

```python
print(stati.mode(["yes","yes","no","no","no",]))
```

```python
print(stati.mode(["yellow","green","pink","yellow","red"]))
```

a distribution with more then one mode is called multimode

```python
# qualitative data

from statistics import multimode
values=["yes","yes","yes","no","no","no",]
x=multimode(values)
print("x",x)

# quantitative data

values=[1,1,1,2,2,3,4,5,5,5]
y=multimode(values)
print("y",y)
```
