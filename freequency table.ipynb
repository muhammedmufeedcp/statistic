```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create sample data
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'B', 'C', 'B', 'C', 'A', 'C', 'B', 'A', 'C'],
    'Scores': [85, 90, 88, 75, 95, 88, 90, 78, 88, 80, 90, 78, 92, 85, 77],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female']
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
'''Frequency means the number of times a value appears in the data.
  A table can quickly show us how many times each value appears'''
```

```python
# frequancy table for "category"
category_freq=df["Category"].value_counts()
print("frequancy table for category:\n",category_freq)
```

```python
# plot for "category"
sns.countplot(x="Category",data=df,palette="magma")
plt.title("frequancy distribution : category")
plt.xlabel("category")
plt.ylabel("frequency")
```

```python
Gender_freq=df["Gender"].value_counts()
print("frequancy table for Gender:\n",Gender_freq)
```

```python
# plote of 'gender'
sns.countplot(x="Gender",data=df,palette="cool")
plt.title("frequancy distribution : Gender")
plt.xlabel("Gender")
plt.ylabel("frequency")
```

#frequancy table for numerical data

```python
# frequancy table for "scores"
Scores_freq=df["Scores"].value_counts()
print("frequency table for 'scores' ; \n",Scores_freq)
```

```python
# plot score
sns.countplot(x="Scores",data=df,palette='magma')
plt.title("frequency distribution : score")
plt.xlabel("Scores")
plt.ylabel("frequency")
```

```python
'''if the data has many diffrent values,it is easier
 to use intarvales of values to presentthrm in a table'''
```

```python
# creating bins for 'scores'
df["Scores_bins"]=pd.cut(df["Scores"],bins=[70,80,90,100],labels=['70-80','80-90','90-100'])
# frequancy table for 'score_bins'
score_bin_freq=df["Scores_bins"].value_counts().sort_index()
print("\nfrequuency table for 'Score_bins :\n",score_bin_freq)
```

```python
sns.countplot(x="Scores_bins",data=df,palette='cool',legend=True,hue='Scores_bins')
plt.title("frequency distribution : Score")
plt.xlabel("Scores Range")
plt.ylabel("frequency")
plt.show()
```
