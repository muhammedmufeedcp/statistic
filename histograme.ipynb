```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the iris dataset
from sklearn.datasets import load_iris
iris=load_iris()
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df
```

this data sets consists of 3 diffrent types of iris flowers(setosa,vericolour,and virginca)petal and sepal length

#histograms

```python
plt.hist(iris_df['sepal length (cm)'],color='lightblue',edgecolor="black")
plt.title("histograme of sepal length (cm)")
```

```python
# distributhion
sns.histplot(iris_df["sepal length (cm)"],kde=True)
```

```python
plt.hist(iris_df['petal length (cm)'],color='red',edgecolor="black")
plt.title("histograme of sepal length (cm)")
```

```python
sns.histplot(iris_df["petal length (cm)"],kde=True)
```

```python
iris_df.hist(edgecolor='red')
plt.suptitle("histogram of iris dataset features",fontsize=15)
plt.show()
```

Here are some inferences that can be drawn from the histograms for the Iris dataset:

### **1. Sepal Length**:
- The data appears to be somewhat normally distributed, but with a slight right skew. The majority of the values are clustered between 5 and 6 cm.


### **2. Sepal Width**:
- The distribution is not normal; it has two prominent peaks, indicating a **bimodal distribution**. The data points are spread more broadly, ranging from 2 cm to 4 cm.


### **3. Petal Length**:
- This feature shows a **strong right-skew**. The majority of the data points are concentrated on the lower end (around 1 to 2 cm), with fewer data points in the higher range.


### **4. Petal Width**:
- Similar to petal length, the distribution is **right-skewed**, with most values clustered between 0.1 cm and 0.6 cm.


### **Overall Insights**:
- **Skewness**: Both petal length and petal width are right-skewed, indicating that most Iris flowers in the dataset have shorter petals.
- **Bimodal distribution**: Sepal width has a bimodal distribution, which might indicate that different species of Iris exhibit different sepal widths.
