```python
# importing necessary libraryies
import numpy as np
import pandas as pd

from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
# example dataset : actual vs predicted value
actual=np.array([3,-0.5,2,7])
predicted=np.array([2.5,0.0,2,9])
```

```python
# creating a dataframe
df = pd.DataFrame({"actual":actual,
                   "predicted": predicted})
print(df)
```

#mean absolute error-MAE

```python
# calculaiting mean absolute error (MAE) using the function
mae=mean_absolute_error(actual,predicted)
print("mean absolute error",mae)
```

# mean squared error-MSE

```python
# calculaiting mean square error (MSE) using the function
mse=mean_squared_error(actual,predicted)
print("mean squared error (MSE):",mse)
```
