```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#wine data

```python
from sklearn.datasets import load_wine

wine_data=load_wine(as_frame=True)
wine_df=wine_data.frame
wine_df.head()
```

#coveriance

```python
covariance_metrix = wine_df.cov()
covariance_metrix
```

#corrilation

```python
wine_correlation_metrix = wine_df.corr()
wine_correlation_metrix
```

#heatmap

```python
plt.figure(figsize=(12, 8))
sns.heatmap(wine_correlation_metrix,annot=True, cmap='coolwarm')
plt.title('correlation heatmap')
plt.show()
```

#iris data

```python
from sklearn.datasets import load_iris

iris_data=load_iris(as_frame=True)
iris_df=iris_data.frame
iris_df.head()
```

```python
covariance_metrix = iris_df.cov()
covariance_metrix
```

```python
iris_correlation_metrix = iris_df.corr()
iris_correlation_metrix
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(iris_correlation_metrix,annot=True, cmap='coolwarm')
plt.title('correlation heatmap')
plt.show()
```

#scatter plot

```python
# Scatter plot between Sepal Length and Sepal Width, colored by target (class labels)
plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c=iris_df['target'], cmap='viridis')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot: Sepal Length vs Sepal Width (Colored by Target)')
plt.colorbar(label='Target')
plt.show()
```

```python
'''**Scatter Plot Inferences:**


1. Cluster Separation:
* When visualizing Sepal Length vs. Sepal Width and Petal Length vs. Petal Width, the scatter plots reveal clear separation between the three species (color-coded by the target variable).
* Setosa: This species tends to cluster tightly in its own space for both sepal and petal features, indicating that its easily distinguishable from the other species.
* Versicolor and Virginica: These two species have overlapping clusters, particularly in the sepal feature space, but they are more distinguishable in the petal feature space (especially in petal length and width).
2. Strong Linear Relationship:
The scatter plots between Petal Length and Petal Width show a clear linear relationship for each species, especially for Versicolor and Virginica. This reinforces the high positive correlation seen in the heatmap.'''
```
