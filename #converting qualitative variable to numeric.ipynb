#dataframe print

```python
# importing the necessary libraries
import pandas as pd
```

```python
data = {
    'name': ['Sheldon', 'Penny', 'Amy', 'Penny', 'Raj', 'Sheldon'],
    'age': [42, 24, 31, 29, 37, 40],
    'gender': ['male', 'female', 'female', 'female', 'male', 'male'],
    'education': ['Under-Graduate', 'Diploma ' , 'Post-Graduate', 'Diploma ', 'Under-Graduate', 'Post-Graduate'],
    'place' : ['malappuram', 'calicut', 'kochi', 'malappuram', 'calicut', 'kochi']
    }
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
df.dtypes
```

'name', 'gender', 'place' and 'education' are Qualitative Variables. Machine Learning models can not work on categorical(qualitative) variables as strings, so we need to change them into numerical form.

# method 1:using replace()method



```python
df1=df
df1['gender'].replace(['male','female'],[0,1],inplace=True)
```

```python
df1
```

# method 2:using get_dummies()

This method will return the dummy variable columns. In this case, in the education variable we have 3 different values so, it returned three columns

```python
df["education"].unique()
```

```python
# get the dummies and store it in a variable
dummies=pd.get_dummies(df["education"])
dummies
```

```python
# get the dummies with integers value and store it in a variable
dummies=pd.get_dummies(df["education"]).astype(int)
dummies
```

```python
# concatenate the dummies to original dataframe
df2=pd.concat([df,dummies],axis="columns")

df2
```

```python
#drop the original variable
df2.drop(["education"],axis="columns",inplace=True)

#print the dataframe
print(df2)
```

# method 3:label encoding

```python
# importing label encoder from sklearn library from preprocessing module
from sklearn.preprocessing import LabelEncoder

#creating a instance of lebel encounder
le=LabelEncoder()
```

```python
df3=df2
df3["name"]=le.fit_transform(df2["name"])
print(df3)
```
