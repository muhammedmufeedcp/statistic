```python
# import necessary libraries

import numpy as np
import pandas as pd
import statistics as stati
```

# data

```python
# people data
data=[50,60,55,65,70]
```

```python
# step 1: calculate the mean
mean=np.mean(data)
print("population = sample mean= ",mean)
```

# sample variance

```python
var=stati.variance(data)
print("the sample variance :",var)
```

```python
stdv=stati.stdev(data)
print("standard deviation is :", stdv)
```

# the population variance

```python
from statistics import pvariance
```

```python
pvar=stati.pvariance(data)
print("the population variance is :",pvar)
```

```python
pvar=stati.pstdev(data)
print("the population standerd deviation is :",pvar)
```
