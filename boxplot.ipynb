```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the iris dataset
from sklearn.datasets import load_iris
iris=load_iris()
iris_df=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df
```

```python
print(np.quantile(iris_df['sepal length (cm)'],[0.25,0.5,0.75]))
# 0.25 meas q1(25%)
# 0.5 mans q2/median(50%)
# 0.75 means q3(75%)
```

```python
plt.boxplot(iris_df['sepal length (cm)'], vert=False, patch_artist=True
            )
plt.title('boxplot of sepal length')
plt.xlabel('value')
```

```python
plt.boxplot(iris_df['sepal width (cm)'], vert=False, patch_artist=True )
plt.title('boxplot of sepal width (cm)')
plt.xlabel('value')
```

```python
# plotting boxplot for each feature
iris_df.plot(kind='box',subplots=True, layout=(1, 4), figsize=(18, 6),
             color='red')
plt.suptitle('boxplots of iris dataset features' , fontsize=16)
plt.show()
```

```python
iris_df.describe()
```

```python
# load the wine dataset
from sklearn.datasets import load_wine
wine=load_wine()
wine_df=pd.DataFrame(wine.data,columns=wine.feature_names)
wine_df
```

```python
wine_df.plot(kind='box',subplots=True, layout=(3, 5), figsize=(18, 6),
             color='red')
plt.suptitle('boxplots of iris dataset features' , fontsize=16)
plt.show()
```
